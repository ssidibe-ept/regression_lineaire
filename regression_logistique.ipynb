```python
##Importation des bibliotheques
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

1) Charger les données

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

#Affichage de l'age en fonction du target

```python
plt.scatter(df['age'], df['target'])
plt.xlabel('Age')
plt.ylabel('Target')
plt.title('Age vs Target')
```

#Affichage de target en fonction de tension

```python
plt.scatter(df['trestbps'], df['target'])
plt.xlabel('Tension')
plt.ylabel('Target')
plt.title('Tension vs Target')
```

#Affichage de target en fonction de chol

```python
plt.scatter(df['chol'], df['target'])
plt.xlabel('Chol')
plt.ylabel('Target')
plt.title('Tension vs Target')
```
