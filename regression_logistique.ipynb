```python
##Importation des bibliotheques
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import  seaborn as sns
```

1) Charger les données

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

#Affichage de l'age en fonction du target

```python
plt.scatter(df['age'], df['target'])
plt.xlabel('Age')
plt.ylabel('Target')
plt.title('Age vs Target')
```

#Affichage de target en fonction de tension

```python
plt.scatter(df['trestbps'], df['target'])
plt.xlabel('Tension')
plt.ylabel('Target')
plt.title('Tension vs Target')
```

```python
sns.pairplot(df, hue='target')
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
print(df["target"].value_counts())
```

```python
malade=df[df["target"]==1]
```

```python
malade.head()
```

```python
non_malade=df[df["target"]==0]
```

```python
non_malade.head()
```

```python
plt.scatter(malade["trestbps"],malade["target"], color="red", marker="x",label="malade")
plt.scatter(non_malade["trestbps"],non_malade["target"], color="blue",marker="+",label="non_malade")
plt.xlabel("trestbps")
plt.ylabel("target")
plt.legend()
plt.title("courbe des malades et non_malades")
plt.show()
```
